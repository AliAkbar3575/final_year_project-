```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
import pickle
```

```python
plt.figure(figsize=(12,7))

plt.style.use("_mpl-gallery")

name = ['ResNet', 'VGGNet', 'EfficientNet','DenseNet', 'MobileNet', 'XceptionNet']

x = np.arange(len(name))+1
accuracy1 = np.array([95.77,94.63,93.34,92.50,90.01,92.72])
accuracy2 = np.array([94.83, 97.39, 97.10, 89.43, 94.87, 91.35])

width = 0.25
plt.bar(x, accuracy1, width=width, color="#003366", label="pre-trained")
plt.bar(x+width+0.08, accuracy2, width=width, color="#FF3333", label="modified")

plt.xlabel("Model", fontsize=20)
plt.ylabel("accuracy", fontsize=20)
plt.ylim(85,100)
plt.xticks(x+0.17,name)

plt.tick_params(axis='both', which='major', labelsize=12)

plt.legend(fontsize=13)
# Save the silu_ to a file (e.g., PNG format) with high resolution
plt.savefig('images1/comparison11.png', dpi=600)  # Adjust DPI as needed
plt.show()
```

```python
accu_hist = pickle.load(open("accu_hist", "rb"))
loss_hist = pickle.load(open("loss_hist", "rb"))

plt.figure(figsize=(16,10))
# plt.style.use("_mpl-gallery")
epochs = np.arange(25)+1
for i in range(len(accu_hist)):
    
    plt.subplot(2,3,i+1)
    plt.plot(epochs, np.array(loss_hist[i])*100, linewidth=2, color="#003366")
    plt.xlabel("epochs", fontsize=12)
    plt.ylabel("loss(%)", fontsize=12)
    plt.title(f"fold-0{i+1}")
    plt.tick_params(axis='both', which='major', labelsize=9)


# plt.legend()
# Save the silu_ to a file (e.g., PNG format) with high resolution
plt.savefig('images1/loss_hist.png', dpi=400)  # Adjust DPI as needed
plt.show()
```

```python
cm = pickle.load(open("con_mat.pkl","rb"))

merged_cm = np.zeros((3,3))
for i in cm:
    merged_cm = merged_cm + i

plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)

sns.heatmap(np.round(merged_cm/5).astype(int), annot=True, fmt="d", cmap="Blues", linewidths=.5, cbar=False, square=True,
            xticklabels=['CAP', 'COVID', 'non-COVID'], yticklabels=['CAP', 'COVID', 'non-COVID'])

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')

plt.show()
plt.savefig('images1/cm.png', dpi=600, bbox_inches='tight')
```
